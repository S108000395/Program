```python
import sqlite3
import requests
import pandas as pd

number = '2303';

def get_financial_statements(number):

    url_1 = 'https://mops.twse.com.tw/mops/web/ajax_t164sb03'
    url_2 = 'https://mops.twse.com.tw/mops/web/ajax_t164sb04'

    payload = {
            'encodeURIComponent':'1',
            'step':'1',
            'firstin':'1',
            'off':'1',
            'queryName':'co_id',
            'inpuType':'co_id',
            'TYPEK':'all',
            'isnew':'true',
            'co_id': str(number)
    }

    res = requests.post(url_1, data = payload)
    df1 = pd.read_html(res.text)[1]
    df1.to_csv("D://"+ number + "_1.csv", encoding='utf-8-sig')

    payload = {
            'encodeURIComponent':'1',
            'step':'1',
            'firstin':'1',
            'off':'1',
            'queryName':'co_id',
            'inpuType':'co_id',
            'TYPEK':'all',
            'isnew':'true',
            'co_id': str(number)
    }

    res = requests.post(url_2, data = payload)
    df2 = pd.read_html(res.text)[1]
    df2.to_csv("D://" + number + "_2.csv", encoding='utf-8-sig')

    return df1,df2
    
get_financial_statements(number)
```
